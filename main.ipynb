## Arsitektur Multi-Agent AI

```text
                         User
                           │
                           ▼
                 Supervisor Agent
                           │
        ┌──────────────────┼──────────────────┐
        ▼                  ▼                  ▼
    CV Agent       Interview Agent    Training Agent
        │                  │                  │
        └──────────────────┼──────────────────┘
                           │
                           ▼
              Client Relations Agent
                           │
                           ▼
                  Reporting Agent
                           │
                           ▼
                         User
```

### Alur Kerja

1. **User** mengirimkan permintaan atau pertanyaan terkait proses rekrutmen.
2. **Supervisor Agent** menerima permintaan dan menentukan agent yang paling sesuai untuk menangani tugas.
3. **CV Agent** melakukan analisis CV dan pencocokan kandidat dengan kebutuhan posisi.
4. **Interview Agent** menyusun pertanyaan interview serta mengevaluasi hasil interview kandidat.
5. **Training Agent** memberikan rekomendasi pelatihan atau pengembangan kompetensi kandidat.
6. Hasil dari seluruh agent dikirim ke **Client Relations Agent** untuk disesuaikan dengan kebutuhan perusahaan atau klien.
7. **Reporting Agent** menyusun laporan akhir berdasarkan seluruh hasil analisis.
8. Laporan dikembalikan kepada **User** sebagai output sistem.

````python
# Membuat Embedding Model
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embedding Model Loaded")
````

````python
# Membuat Vector Database (FAISS)
vectorstore = FAISS.from_documents(
    documents=chunks,
    embedding=embedding_model
)

print("FAISS Database Berhasil Dibuat")
````

````python
# Membuat Fungsi Mengambil Context
def retrieve_context(question: str):

    docs = retriever.invoke(question)

    context = "\n\n".join(
        [doc.page_content for doc in docs]
    )

    return context, docs
````

````python
# Membuat Fungsi RAG
def rag_answer(question: str):

    context, docs = retrieve_context(question)

    prompt = rag_prompt.format_messages(
        context=context,
        question=question
    )

    response = llm.invoke(prompt)

    return {
        "answer": response.content,
        "documents": docs
    }
````

````python
# Import Library Agent
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()

print("Agent Library Loaded")
````

````python
# Base Function Agent
# Semua agent akan memakai fungsi ini supaya tidak ada kode yang berulang.
def run_agent(role: str, question: str):
    """
    Menjalankan agent berdasarkan role yang diberikan.
    Semua agent menggunakan RAG Tool yang sama.
    """

    rag_result = rag_answer(question)

    context = "\n\n".join(
        [doc.page_content for doc in rag_result["documents"]]
    )

    prompt = ChatPromptTemplate.from_template(
        """
Anda adalah {role} pada perusahaan HR Recruitment.

Tugas Anda:

- Jawab hanya berdasarkan context.
- Jangan mengarang jawaban.
- Jika informasi tidak tersedia, katakan informasi tidak ditemukan.
- Berikan jawaban yang profesional.

Context:
{context}

Question:
{question}
"""
    )

    chain = prompt | llm | output_parser

    answer = chain.invoke({
        "role": role,
        "context": context,
        "question": question
    })

    return {
        "agent": role,
        "answer": answer,
        "documents": rag_result["documents"]
    }
````

````python
# CV Screening Agent
def cv_agent(question):
    return run_agent(
        role="CV Screening Specialist",
        question=question
    )
````

````python
# Interview Agent
def interview_agent(question):
    return run_agent(
        role="Interview Coordinator",
        question=question
    )
````

````python
# Training Agent
def training_agent(question):
    return run_agent(
        role="Training and Onboarding Specialist",
        question=question
    )
````

````python
# Client Relations Agent
def client_agent(question):
    return run_agent(
        role="Client Relations Officer",
        question=question
    )
````

````python
# Reporting Agent
def reporting_agent(question):
    return run_agent(
        role="Reporting Analyst",
        question=question
    )
````

# PART 5: Supervisor Agent

````python
# Daftar Agent
AGENTS = {
    "cv": cv_agent,
    "interview": interview_agent,
    "training": training_agent,
    "client": client_agent,
    "report": reporting_agent
}

print(AGENTS.keys())
````

````python
# Supervisor Prompt
supervisor_prompt = ChatPromptTemplate.from_template("""
Anda adalah Supervisor Agent.

Tugas Anda adalah memilih agent yang PALING SESUAI.

Pilihan agent hanya:

cv
interview
training
client
report

Aturan:

- Pertanyaan tentang CV → cv
- Pertanyaan tentang screening → cv
- Pertanyaan tentang interview → interview
- Pertanyaan onboarding → training
- Pertanyaan training → training
- Pertanyaan client → client
- Pertanyaan progress → report
- Pertanyaan laporan → report
- Pertanyaan ringkasan → report

Jawab SATU KATA SAJA.

Question:
{question}
""")
````

````python
# Supervisor Chain
supervisor_chain = (
    supervisor_prompt
    | llm
    | output_parser
)

print("Supervisor Ready")
````

````python
# Supervisor Agent
def supervisor_agent(question):

    q = question.lower()

    # CV Agent
    if any(keyword in q for keyword in [
        "cv",
        "screening",
        "resume"
    ]):
        selected = "cv"

    # Interview Agent
    elif any(keyword in q for keyword in [
        "interview",
        "hr interview",
        "user interview"
    ]):
        selected = "interview"

    # Training Agent
    elif any(keyword in q for keyword in [
        "training",
        "onboarding",
        "orientasi"
    ]):
        selected = "training"

    # Client Relations Agent
    elif any(keyword in q for keyword in [
        "client",
        "klien"
    ]):
        selected = "client"

    # Reporting Agent
    elif any(keyword in q for keyword in [
        "laporan",
        "report",
        "ringkasan",
        "proses recruitment",
        "tahapan recruitment",
        "workflow"
    ]):
        selected = "report"

    # Jika tidak cocok dengan rule, baru gunakan LLM
    else:
        selected = supervisor_chain.invoke({
            "question": question
        }).strip().lower()

        if selected not in AGENTS:
            selected = "report"

    result = AGENTS[selected](question)
    result["selected_agent"] = selected

    return result
````

# PART 6 — Evaluation

````python
# Dataset Pengujian
# Buat pertanyaan acuan beserta agent yang seharusnya dipilih.
test_cases = [
    {
        "question": "Apa saja tahapan proses recruitment?",
        "expected_agent": "report",
        "expected_keyword": [
            "recruitment",
            "screening",
            "interview",
            "offering",
            "onboarding"
        ]
    },
    {
        "question": "Apa saja kriteria screening CV?",
        "expected_agent": "cv",
        "expected_keyword": [
            "cv",
            "kandidat",
            "pengalaman",
            "pendidikan"
        ]
    },
    {
        "question": "Bagaimana proses interview kandidat?",
        "expected_agent": "interview",
        "expected_keyword": [
            "interview",
            "hr",
            "user"
        ]
    },
    {
        "question": "Bagaimana proses onboarding?",
        "expected_agent": "training",
        "expected_keyword": [
            "onboarding",
            "training"
        ]
    },
    {
        "question": "Bagaimana update status kandidat kepada client?",
        "expected_agent": "client",
        "expected_keyword": [
            "client",
            "status"
        ]
    }
]

print("Jumlah Test Case :", len(test_cases))
````

````python
# Accuracy (Routing Agent)
# Yang diuji adalah apakah Supervisor memilih agent yang benar.
correct = 0

for tc in test_cases:

    result = supervisor_agent(tc["question"])

    predicted = result["selected_agent"]

    print("=" * 50)
    print("Question :", tc["question"])
    print("Expected :", tc["expected_agent"])
    print("Predicted:", predicted)

    if predicted == tc["expected_agent"]:
        correct += 1

accuracy = correct / len(test_cases)

print("=" * 50)
print(f"Accuracy : {accuracy:.2%}")
````

````python
# Effectiveness
# Effectiveness Evaluation
# Effectiveness Evaluation
from rapidfuzz import fuzz

effectiveness_scores = []

print("=" * 60)
print("Effectiveness Evaluation")
print("=" * 60)

for tc in test_cases:

    result = supervisor_agent(tc["question"])
    answer = result["answer"].lower()

    keyword_scores = []

    for keyword in tc["expected_keyword"]:

        score = fuzz.partial_ratio(
            keyword.lower(),
            answer
        )

        keyword_scores.append(score)

    avg_score = sum(keyword_scores) / len(keyword_scores)

    effectiveness_scores.append(avg_score)

    print("=" * 60)
    print("Question :", tc["question"])
    print("Keyword Scores :", keyword_scores)
    print(f"Effectiveness : {avg_score:.2f}%")

average_effectiveness = (
    sum(effectiveness_scores) /
    len(effectiveness_scores)
)

print("=" * 60)
print(f"Average Effectiveness : {average_effectiveness:.2f}%")
````

````python
# Efficiency
import time

response_times = []

print("=" * 60)
print("Efficiency Test")
print("=" * 60)

for tc in test_cases:

    start = time.perf_counter()

    supervisor_agent(tc["question"])

    end = time.perf_counter()

    elapsed = end - start

    response_times.append(elapsed)

    print(f"{tc['question']}")
    print(f"Response Time : {elapsed:.3f} detik\n")

average_time = sum(response_times) / len(response_times)

print("=" * 60)
print(f"Average Response Time : {average_time:.3f} detik")
````
